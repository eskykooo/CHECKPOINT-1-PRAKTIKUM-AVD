# **MELAKUKAN IMPORT**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "/content/data_mental_fix.csv"
df = pd.read_csv(file_path)
df
```

# **ANALISIS DAN VISUALISASI DATA**

## DISTRIBUSI OCCUPATION

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Occupation", data=df, hue="Occupation", legend=False, palette="muted")
plt.title("Distribusi Pekerjaan")
plt.xlabel("Jenis Pekerjaan")
plt.ylabel("Jumlah")
plt.xticks(rotation=45)
plt.show()
```

### INSIGHT

Terlihat bahwa kategori mahasiswa (student) dan ibu rumah tangga (housewife) memiliki jumlah tertinggi dibandingkan kategori lainnya. Hal ini menunjukkan bahwa kedua kelompok lebih sering mengalami permasalahan terkait kesehatan mental. Sebaliknya, kategori business memiliki jumlah terendah, yang dapat mengindikasikan bahwa individu di bidang bisnis kurang terwakili dalam data atau memiliki keterbatasan dalam mengakses layanan kesehatan mental.

### TINDAKAN

Berdasarkan temuan ini, beberapa tindakan dapat diambil untuk meningkatkan kesadaran dan dukungan terhadap kesehatan mental. Pertama, mahasiswa dan ibu rumah tangga dapat menjadi fokus utama dalam penyediaan program dukungan, seperti layanan konseling atau pelatihan manajemen stres yang disesuaikan dengan tantangan mereka. Kedua, perlu ada upaya untuk meningkatkan aksesibilitas layanan kesehatan mental bagi pelaku bisnis, misalnya melalui kampanye edukasi atau penyediaan layanan fleksibel yang sesuai dengan jadwal mereka.

## DISTRIBUSI GROWING STRESS

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Growing_Stress", data=df, hue="Growing_Stress", legend=False, palette="muted")
plt.title("Distribusi Peningkatan Stres")
plt.xlabel("Peningkatan Stres")
plt.ylabel("Jumlah")
plt.show()
```

Dari visualisasi distribusi peningkatan stres di atas, mayoritas responden mengaku mengalami peningkatan stres (Yes), sementara sebagian lainnya tidak mengalami peningkatan stres (No) atau merasa ragu-ragu (Maybe). Hal ini menunjukkan bahwa stres merupakan masalah yang signifikan bagi banyak individu, dengan sebagian besar responden menyadari adanya peningkatan tingkat stres dalam kehidupan mereka. Sementara itu, jumlah responden yang menjawab Maybe cukup tinggi, yang mungkin mengindikasikan kurangnya kesadaran atau pemahaman mereka terhadap gejala stres yang mereka alami.

### TINDAKAN

Untuk mengatasi permasalahan ini, beberapa tindakan dapat dilakukan. Pertama, penting untuk meningkatkan kesadaran masyarakat mengenai tanda-tanda stres dan dampaknya terhadap kesehatan mental, misalnya melalui kampanye edukasi atau seminar kesehatan mental. Kedua, penyediaan akses ke layanan dukungan psikologis, seperti konseling atau terapi, dapat membantu mereka yang mengalami peningkatan stres agar dapat mengelola kondisi mereka dengan lebih baik.

## HUBUNGAN GROWING STRESS DENGAN OCCUPATION

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Growing_Stress'].value_counts()
    for x_label, grp in df.groupby('Occupation')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Occupation')
_ = plt.ylabel('Growing_Stress')
```

### INSIGHT

Dari heatmap yang ditampilkan, terlihat bahwa tingkat stres yang meningkat banyak dialami oleh mahasiswa (Student), pekerja kantoran (Corporate), dan ibu rumah tangga (Housewife). Mahasiswa mungkin mengalami tekanan dari tuntutan akademik, jadwal yang padat, serta ketidakpastian masa depan. Pekerja kantoran berpotensi mengalami stres karena beban kerja yang tinggi, tekanan dari atasan, dan keseimbangan kehidupan kerja yang kurang optimal. Sementara itu, ibu rumah tangga bisa mengalami stres akibat tanggung jawab domestik yang besar serta kurangnya dukungan sosial yang memadai.

### TINDAKAN

 Tindakan yang dapat diambil antara lain mahasiswa dapat diberikan akses lebih mudah ke layanan konseling kampus serta program manajemen waktu yang lebih baik. Perusahaan perlu mengadopsi kebijakan kesejahteraan karyawan, seperti fleksibilitas kerja dan program kesehatan mental untuk mengurangi stres pekerja kantoran. Sementara itu, bagi ibu rumah tangga, komunitas dukungan atau program berbasis masyarakat bisa dibentuk untuk memberikan bantuan serta ruang untuk berbagi pengalaman.

## HUBUNGAN GROWING STRESS DENGAN QUARANTINE FRUSTRATIONS

Sebaliknya, individu yang tidak merasa frustrasi selama karantina memiliki tingkat stres yang lebih rendah, yang ditunjukkan oleh intensitas warna yang lebih rendah pada bagian "No".

### TINDAKAN

Tindakan yang dapat diambil untuk mengatasi permasalahan ini adalah dengan meningkatkan dukungan sosial dan akses terhadap aktivitas yang dapat mengurangi stres selama masa karantina. Program kesehatan mental seperti sesi terapi daring, kegiatan komunitas virtual, serta promosi gaya hidup sehat dapat membantu mengurangi frustrasi dan tekanan psikologis.
